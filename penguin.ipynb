## Feature Engineering

```python
df["island"]
```

```python
X = pd.get_dummies(df.drop("species", axis=1), drop_first=True) * 1
y = df["species"]
```

```python
X
```

## Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Decision Tree Classifier
## Hyperparameters

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
base_preds = model.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, base_preds))
```

```python
confusion_matrix(y_test, base_preds)
```

```python
pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=["Feature Importance"]).sort_values("Feature Importance")
```

## Visualize the Tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12, 8), dpi=200)
plot_tree(model, filled=True);
```

### Visualization function

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print("\n")
    plt.figure(figsize=(12, 8), dpi=200)
    plot_tree(model, filled=True)
```

## Hyperparameters
### Maximum Depth

```python
pruned_tree = DecisionTreeClassifier(max_depth=3)
```

```python
pruned_tree.fit(X_train, y_train)
```

```python
report_model(pruned_tree)
```

### Maximum Leaf Nodes

```python
leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
leaf_tree.fit(X_train, y_train)
```

```python
report_model(leaf_tree)
```

## Entropy Criterion

```python
entropy_tree = DecisionTreeClassifier(criterion="entropy")
```

```python
entropy_tree.fit(X_train, y_train)
```

```python
report_model(entropy_tree)
```

```python
y_pred = entropy_tree.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```
